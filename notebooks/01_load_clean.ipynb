# 01 - Load and Clean

Raw veriyi oku, temizle ve islenmis dosyayi kaydet.

```python
from pathlib import Path
import sys

import pandas as pd

project_root = Path('..')
sys.path.append(str(project_root / 'src'))

from preprocess import load_raw, clean_data, feature_engineering, save_processed

raw_path = project_root / 'data' / 'raw' / 'SampleSuperstore.csv'
processed_path = project_root / 'data' / 'processed' / 'clean.csv'

df = load_raw(raw_path)
df = clean_data(df)
df = feature_engineering(df)
save_processed(df, processed_path)

missing = df.isna().sum().to_frame('missing')
missing['missing_pct'] = (missing['missing'] / len(df) * 100).round(2)
missing.sort_values('missing', ascending=False)
```
