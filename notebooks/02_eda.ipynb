# 02 - EDA

Histogram, scatter ve heatmap uret.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

project_root = Path('..')
data_path = project_root / 'data' / 'processed' / 'clean.csv'
fig_dir = project_root / 'reports' / 'figures'
fig_dir.mkdir(parents=True, exist_ok=True)

df = pd.read_csv(data_path)

if 'Sales' in df.columns:
    df['Sales'].hist(bins=30)
    plt.title('Histogram: Sales')
    plt.tight_layout()
    plt.savefig(fig_dir / 'hist_sales.png', dpi=150)
    plt.close()

if 'Profit' in df.columns:
    df['Profit'].hist(bins=30)
    plt.title('Histogram: Profit')
    plt.tight_layout()
    plt.savefig(fig_dir / 'hist_profit.png', dpi=150)
    plt.close()

if {'Sales', 'Profit'}.issubset(df.columns):
    plt.scatter(df['Sales'], df['Profit'], alpha=0.4)
    plt.xlabel('Sales')
    plt.ylabel('Profit')
    plt.title('Scatter: Sales vs Profit')
    plt.tight_layout()
    plt.savefig(fig_dir / 'scatter_sales_profit.png', dpi=150)
    plt.close()

num_df = df.select_dtypes(include='number')
if num_df.shape[1] >= 2:
    corr = num_df.corr()
    plt.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
    plt.colorbar()
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.title('Correlation Heatmap')
    plt.tight_layout()
    plt.savefig(fig_dir / 'heatmap_corr.png', dpi=150)
    plt.close()
```

Yorum: Sales ve Profit histogramlari sag kuyruklu; az sayida buyuk siparis dagilimi belirgin sekilde etkiliyor. Sales ile Profit arasinda genel bir birlikte hareket var ama dagilim genis, bu da indirim ve maliyet etkisini hissettiriyor. Korelasyon haritasinda Sales ve Quantity ile Profit arasinda orta duzey bir bag goruluyor.
